## Basic matplotlib exercises

#### Import matplotlib and numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Identify x=[0,6,7],y=[9,10,11], give names to x ;y axes, to the label and title , show legend

```python
x = [0,6,7]
y = [9,10,11]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('The Graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend('legend')
```

```python
x16 = np.arange(1, 5)
# Plot green line with x16
plt.plot(x16)
# Plot red line with x16+1
plt.plot(x16+1)
# Plot blue line with x16+2
plt.plot(x16+2)

```

## Control line styles

All the available line styles are available in the following table:


**Style abbreviation**  **Style**

-                    solid line
   
--                   dashed line
   
-.                   dash-dot line
   
:                    dotted line
   


Now, we can see the default format string for a single line plot is 'b-'.

```python
x16 = np.arange(1, 5)
# plot blue line with -- , red line with -. , green line with :
plt.plot(x16, color = 'b', ls ='--')
plt.plot(x16 + 1, color = 'r', ls='-')
plt.plot(x16 + 2, color = 'g', ls=':')
```

```python
x = np.linspace(0, 2)
# plot x and x and label it as 'linear'
# plot x and x**2 and label it as 'quadratic'
# plot x and x**3 and label it as cubic
# label x and y axes
# show legend
plt.plot(x,x)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.legend(['linear','quadratic','cubic'])
```
